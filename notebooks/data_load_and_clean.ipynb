```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_RAW = Path("../data/raw")
csv_path = DATA_RAW / "atp_tennis.csv"

df = pd.read_csv(csv_path)

# Date Parsing and Cleanup
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Surface"] = df["Surface"].astype(str).str.strip()
df["Round"] = df["Round"].astype(str).str.strip()

num_cols = ["Rank_1","Rank_2","Pts_1","Pts_2","Odd_1","Odd_2","Best of"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

df = df.dropna(subset=["Date","Player_1","Player_2","Winner","Surface"])

print(df.shape)
df.head()
```
